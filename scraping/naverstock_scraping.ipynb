```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import sqlite3 # db에 커넥션
```

```python
# 코스피 시가총액 순위
url = "https://finance.naver.com/sise/sise_market_sum.nhn?&page="

# page2 까지 데이터 추출
for page in range(1,3):
    res = requests.get(url + str(page))
    res.raise_for_status()
    soup = BeautifulSoup(res.text, 'lxml')
    
    # 각각의 행 (1위,2위,,)
    data_rows= soup.select('tbody > tr[onmouseover]')
    for row in data_rows:
        columns = row.select('td')
        data = [column.get_text().strip() for column in columns]
        data.remove('') #'토론실' 데이터 제거
        print(data)
        
        
        #pd.DataFrame(data, columns=['N','종목명','현재가','전일비','등락률','액면가', '시가총액', '상장주식수', '외국인비율','거래량','PER','ROE')
```
