```python
import matplotlib.pyplot as plt

labels = ['Arcade','Action','Adventure','Strategy','Sports'
          ,'Board','Role Playing ','Simulation','Puzzle','Card'
          ,'Educational','Music','Racing', 'Trivia','Casual'
          ,'etc']
sizes = [counts.get('Arcade (Games)'),counts.get('Action (Games)'),counts.get('Adventure (Games)')
                  ,counts.get('Strategy (Games)'),counts.get('Sports (Games)'),counts.get('Board (Games)')
                  ,counts.get('Role Playing (Games)'),counts.get('Simulation (Games)'),counts.get('Puzzle (Games)')
                  ,counts.get('Card (Games)'),counts.get('Educational (Games)'),counts.get('Music (Games)')
                  ,counts.get('Racing (Games)'),counts.get('Trivia (Games)'),counts.get('Casual (Games)')
                  ,etc]

colors = sns.color_palette("Set3", 16)

# colors = ['red','orange', 'yellow', 'yellowgreen','green',
#           'gold', 'lightskyblue', 'lightcoral','pink', '#d395d0',
#           '#8fd9b6', '#66CDAA', '#7FFFD4', '#D2B48C', '#FFA07A'
#          , '#00FF7F']

##  , ,    .
labels_frequency = zip(labels,sizes,colors) 
labels_frequency = sorted(labels_frequency,key=lambda x: x[1],reverse=True)
 
sorted_labels = [x[0] for x in labels_frequency] ##  
sorted_frequency = [x[1] for x in labels_frequency] ##  
sorted_colors = [x[2] for x in labels_frequency] ##  
 
fig = plt.figure(figsize=(15,15)) ##  
fig.set_facecolor('white') ##    
ax = fig.add_subplot() ##  
 
pie = ax.pie(sorted_frequency, ##  
       startangle=90, ##  90(degree) 
       counterclock=False, ##  
       colors = 
             colors, ##  
       )
 
total = np.sum(sizes) ##  
 
threshold = 2.8
sum_pct = 0 ## 
count_less_5pct = 0 ## 5%   
spacing = 0.1
for i,l in enumerate(sorted_labels):
    ang1, ang2 = ax.patches[i].theta1, ax.patches[i].theta2 ##     
    center, r = ax.patches[i].center, ax.patches[i].r ##   
    
    ##      .
    if sorted_frequency[i]/total*100 < threshold:
        x = (r/2+spacing*count_less_5pct)*np.cos(np.pi/180*((ang1+ang2)/2)) + center[0] ##  x
        y = (r/2+spacing*count_less_5pct)*np.sin(np.pi/180*((ang1+ang2)/2)) + center[1] ##  y
        count_less_5pct += 1
    else:
        x = (r/2)*np.cos(np.pi/180*((ang1+ang2)/2)) + center[0] ##  x
        y = (r/2)*np.sin(np.pi/180*((ang1+ang2)/2)) + center[1] ##  y
    
    ##  
    if i < len(labels) - 1:
        sum_pct += float(f'{sorted_frequency[i]/total*100:.2f}')
        ax.text(x,y,f'{sorted_frequency[i]/total*100:.2f}%',ha='center',va='center',fontsize=12)
    else: ##      100   
        ax.text(x,y,f'{100-sum_pct:.2f}%',ha='center',va='center',fontsize=12)

plt.legend(pie[0],sorted_labels, loc='upper right') ## 

fig.set_size_inches(8,8) ##  
plt.show()
```
