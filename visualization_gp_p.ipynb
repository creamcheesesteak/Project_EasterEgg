```python
set_group = ['Games', 'etc']
```

```python
count_group = []
group = []
sum = 0
for group in set_group: # group : key = Life
    name = dict[group]
    group = dict[group] # dict['Life']
    for category in group: # Food and Drink (Applications)
        for key, value in counts.items():
            if key == category:
                sum = value + sum
    count_group.append(sum)
    print(sum)
    sum = 0
```

```python
# importing library and set figure size

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
```

```python
Games = set_group.get('Games')
etc = set_group.get('etc')
```

```python
counts.get('Arcade (Games)')
```

```python
# info. of groups
    
group_names = ['Games', 'etc']
group_sizes = [Games, etc]

# info. of subgroups
subgroup_names = ['Arcade (Games)','Action (Games)','Adventure (Games)','Strategy (Games)','Sports (Games)'
                  ,'Board (Games)','Role Playing (Games)','Simulation (Games)','Puzzle (Games)','Card (Games)'
                  ,'Educational (Games)','Music (Games)','Racing (Games)', 'Trivia (Games)','Casual (Games)'
                  , 'etc']
subgroup_sizes = [counts.get('Arcade (Games)'),counts.get('Action (Games)'),counts.get('Adventure (Games)')
                  ,counts.get('Strategy (Games)'),counts.get('Sports (Games)'),counts.get('Board (Games)')
                  ,counts.get('Role Playing (Games)'),counts.get('Simulation (Games)'),counts.get('Puzzle (Games)')
                  ,counts.get('Card (Games)'),counts.get('Educational (Games)'),counts.get('Music (Games)')
                  ,counts.get('Racing (Games)'),counts.get('Trivia (Games)'),counts.get('Casual (Games)')
                  ,etc]

# colors
a, b = [plt.cm.Reds, plt.cm.Greens]

# width
width_num = 0.4
```

```python
# Outside Ring

fig, ax = plt.subplots()
ax.axis('equal')
pie_outside, _ = ax.pie(group_sizes, 
                        radius=1.3, 
                        labels=group_names, 
                        labeldistance=0.8,
                        colors=[a(1.0), b(0.6)])

plt.setp(pie_outside, 
         width=width_num, 
         edgecolor='white')

# Inside Ring

pie_inside, plt_labels, junk = \
    ax.pie(subgroup_sizes, 
           radius=(1.3 - width_num), 
           labels=subgroup_names, 
           labeldistance=0.75, 
           autopct='%1.1f%%', 
           colors=[a(0.9),a(0.85),a(0.8),a(0.75),a(0.7)
                   ,a(0.65),a(0.6),a(0.55),a(0.5),a(0.45)
                   ,a(0.4),a(0.35),a(0.3),a(0.25),a(0.2)
                   ,b(0.5)])

plt.setp(pie_inside, 
         width=width_num, 
         edgecolor='white')

plt.title('Donut Plot with Subgroups', fontsize=20)
plt.show()
```
